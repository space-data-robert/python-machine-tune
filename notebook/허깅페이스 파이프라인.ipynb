### <font color='red'>추론을 위한 파이프라인 </font>

자연어처리, 컴퓨터비전, 음성 등 학습된 모델을 간단하게 사용할수 있다고 해요.

특징으로는 허깅 페이스에서 모델을 업로드하고 공유할 수 있다고 해요.

- 기존 모델 ```bert``` 뿐만 아니라 ```bert-uncased-fintuned``` 등 기학습된 모델

파이프라인 사용 방식은 간단합니다.

- ```pipeline(태스크 종류, 원하는 태스크 모델)```
- 기학습된 모델 [스페이스](https://huggingface.co/tasks)





````python
from transformers import pipeline
````

````python
sentiment_classfier = pipeline(
    task='text-classification', 
    model='distilbert-base-uncased-finetuned-sst-2-english'
)
````

````python
sentiment_classfier('this movie is so funny.')
````

````python
kor_sentiment_classfier = pipeline(
    task='sentiment-analysis', 
    model='snunlp/KR-FinBert-SC'
)
````

````python
kor_sentiment_classfier([
    '이 상품은 비추해요.',
    '이 상품은 더이상 구매하고 싶지 않습니다.'
])
````

````python
kor_generator = pipeline(
    task='text-generation', 
    model='skt/kogpt2-base-v2'
)
````

````python
sentence = '대교씨엔에는 인공지능 학습에'

kor_generator(
    sentence,
    max_length=100,
    num_return_sequences=2
)
````
